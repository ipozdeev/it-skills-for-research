```python
from sqlalchemy import create_engine
from sqlalchemy import text
import pandas as pd
import numpy as np
```

```python
# Define the connection parameters (as in the Dockerfile)
user = 'postgres'
password = 'drowssap'
db = 'example_db'
host = '0.0.0.0'
port = '5432'

# Create the connection URL
connection_url = f'postgresql+psycopg2://{user}:{password}@{host}:{port}/{db}'
```

```python
# Create the engine
engine = create_engine(connection_url)
```

```python
# insert some data into `time_series_value`
data = pd.DataFrame.from_dict(
    {
        "id": np.arange(1000) + 1,
        "value": np.random.normal(size=(1000, )),
        "date": pd.date_range("2000-01-01 00:00:00", freq="1T", periods=1000),
    }
)

print(data.head())

with engine.connect() as connection:
    data.to_sql(name="time_series_value", con=connection,
                if_exists="append", index=False)
```

```python
# now, let's query some rows
with engine.connect() as connection:
    # Perform database operations
    result = connection.execute(
        text(
            """
            SELECT * 
            FROM time_series_value
            WHERE date < '2000-01-02'::date
            ;
            """
        )
    )
    for row in result:
        print(row)
```
